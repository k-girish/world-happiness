```python
def has_duplicates(x):
    return len(x) > len(x.unique())

def view_all_df(temp):
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        display(temp)
        
def create_rename_dict_from_list_name_tuple(list_name_tuple):
    rename_dict = {}
    for l,n in list_name_tuple:
        for e in l:
            rename_dict[e]=n
    return rename_dict
```
